## Saving the model

```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float32
)

pipe.save_pretrained("./sd_v1_5_local")
```

## Loading model

```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(
    "./sd_v1_5_local",   # local path
    torch_dtype=torch.float32,
    local_files_only=True
)
```

```python
pipe = pipe.to("cuda")
pipe.safety_checker = None

prompt = "A men in white shirt, holding a bat, birds in background, green grass"

image = pipe(prompt, num_inference_steps=25).images[0]
image.save("gpu_result_2.png")
```
